```python
columns_to_fix = [
    "FS_date",
    "LSM",
    "CAP",
    "AST",
    "ALT",
    "T.bil",
    "PLT",
    "PT INR",
    "Alb",
    "Glucose",
    "HbA1c",
    "eGFR",
    "T.chol",
    "HDL",
    "LDL",
    "TG",
    "Weight",
    "Height",
    "BMI"
]

# Apply bfill on a new dataframe
for col in columns_to_fix:
    related_columns = [col] + [f"{col}.{i}" for i in range(1, 6)]  # column, column.1, column.2, ..., column.5
    df[col] = df[related_columns].bfill(axis=1)[col]
```

```python
# Rename and filter columns
columns_to_extract = {
    '(실명)등록번호': 'ID',
    '성별': 'Sex',
    '생년월일': 'Birth Date',
    'Fibrosis stage': 'Fibrosis Stage',
    'NAFLD Type': 'NAFLD Type',
    '사망일자': 'Death Date',
    'Circumference [mm]': 'Waist',#'Waist (허리 둘레 cm)': 'Waist',
    'Last fu_date': 'Last F/U',
    'Smoking': 'Smoking',
    'Ascites': 'Ascites',
    'Ascites_Date': 'Ascites_Date',
    'Varix bleeding': 'VB', 
    'VB_date': 'VB_Date',
    'HEP': 'HEP',
    'HEP_date': 'HEP_Date',
    'HRS': 'HRS',
    'HRS_date': 'HRS_Date',
    'LT': 'LT',
    'LT_date': 'LT_Date',
    'SBP (수축기 혈압) mmHg': 'SBP',
    'DBP (이완기 혈압) mmHg': 'DBP',
    'AVF (Whole Volume)': 'VFA (Volume)',
    'SF (Whole Volume)': 'SFA (Volume)',
    'Spleen (Volume)': 'Spleen (Volume)',
    'AVF (Single Area)': 'VFA (Area)',
    'SF (Single Area)': 'SFA (Area)',
    'Muscle (Single Area)': 'SMA (Area)',
    'AVF (Single HU)': 'VFA (Attenuation)',
    'SF (Single HU)': 'SFA (Attenuation)',
    'Muscle (Single HU)': 'SMA (Attenuation)',
    'Liver/Spleen Volume Ratio': 'Liver/Spleen Volume',
    'Liver/Spleen HU Ratio': 'Liver/Spleen HU',
    'Liver (HU)': 'Liver (HU)',
    'Spleen (HU)': 'Spleen (HU)',
    'Liver Fat (PDFF)': 'Liver (PDFF)'
}

# Select the specified ranges and additional columns
cols_ct_b_to_drinking = df.loc[:, 'CT_B date':'BMI'].columns.tolist()
cols_diabetes_to_nephropathy = df.loc[:, '당뇨병/당뇨병 전단계 유무':'Nephropathy 유무'].columns.tolist()

# Combine all column groups
all_columns = list(columns_to_extract.keys()) + \
              cols_ct_b_to_drinking + \
              cols_diabetes_to_nephropathy
```

```python
# Filter the DataFrame to keep only these columns
filtered_df = df[all_columns]

# Rename specified columns
filtered_df = filtered_df.rename(columns=columns_to_extract)

# Create Death and Death Date columns based on the 사망일자 column
filtered_df['Death'] = filtered_df['Death Date'].apply(lambda x: 0 if x == 0 else 1)
filtered_df['Death Date'] = filtered_df['Death Date'].apply(lambda x: 0 if x == 0 else x)
```

```python
import pandas as pd
import numpy as np

# Step 1: Convert all relevant date columns to datetime format
date_columns = ["CT_B date", "Ascites_Date", "VB_Date", "HEP_Date", "HRS_Date", "LT_Date"]
for col in date_columns:
    filtered_df[col] = pd.to_datetime(filtered_df[col], errors="coerce")

# Step 2: Define a function to find the first LRE event after CT_B date
def find_first_lre(row):
    event_types = {
        1: "Ascites_Date",
        2: "VB_Date",
        3: "HEP_Date",
        4: "HRS_Date",
        5: "LT_Date"
    }
    
    # Store event dates that occur after CT_B date
    valid_events = {
        event_id: row[event_date]
        for event_id, event_date in event_types.items()
        if pd.notna(row[event_date]) and row[event_date] > row["CT_B date"]
    }

    if valid_events:
        # Find the earliest event
        first_event_id = min(valid_events, key=valid_events.get)
        first_event_date = valid_events[first_event_id]
        return first_event_id, first_event_date
    else:
        return 0, 0# No event occurred after CT_B date

# Apply the function to create LRE and LRE Date
filtered_df[["LRE", "LRE Date"]] = filtered_df.apply(lambda row: pd.Series(find_first_lre(row)), axis=1)

# Convert LRE to integer type (for clarity)
filtered_df["LRE"] = filtered_df["LRE"].astype("Int64")  # Keeps NaN as <NA> instead of float NaN
```

```python
# Step 1: Find unique values in the 'NAFLD Type' column
unique_values = filtered_df['NAFLD Type'].unique()
print("Unique values in 'NAFLD Type':", unique_values)

# Step 2: Define the list of values to filter out
X = ['Others (Cancer)', 'NAFL (+)', 'Probable NASH (+)', 'NASH (+)', 'Missing Biopsy']  # Example list of values to remove

# Step 3: Remove rows where 'NAFLD Type' is in the list X
no_cancer_df = filtered_df[~filtered_df['NAFLD Type'].isin(X)]
no_cancer_df = no_cancer_df.reset_index(drop=True)
```

```python
import pandas as pd
import re

# Function to extract only numeric values from a string
def extract_numeric(value):
    if pd.isna(value):  # Keep NaN values unchanged
        return value
    match = re.search(r"[-+]?\d*\.?\d+", str(value))  # Extract number (integer or decimal)
    return float(match.group()) if match else None  # Convert to float if found, else None

fill_columns = [
    "SBP",
    "DBP",
    "VFA (Attenuation)",
    "LSM",
    "CAP",
    "ALT",
    "PT INR",
    "HbA1c",
    "T.chol",
    "HDL",
    "LDL",
    "TG"
]  # Add more column names if needed


# Apply this function to each column in fill_columns
for col in fill_columns:
    no_cancer_na_df[col] = no_cancer_na_df[col].apply(extract_numeric)
```

```python
# Loop through each column in fill_columns and fill NaN with the column median
for col in fill_columns:# Create a new name (e.g., "SBP" from "SBP (수축기 혈압) mmHg")
    no_cancer_na_df[col] = no_cancer_na_df[col].fillna(no_cancer_na_df[col].median())
```
