## 정리
VFA는 (Area) 로 선택  
VFI는 (Volume) 으로 선택하고 키(cm를 m로 변환) 제곱으로 나누기  
SMI는 SMA (Area) 로 선택하고 키("") 제곱으으로 나누기  

```python
# VFI라는 새로운 열을 추가하는데, VFA (Volume)을 (Height / 100) 의 제곱으로 나눈다
data["VFI"] = data["VFA (Volume)"] / (data["Height"] / 100) ** 2

# SFI 도 비슷하게
data["SFI"] = data["SFA (Volume)"] / (data["Height"] / 100) ** 2

# TFI 열을 만들건데 VFI + SFI
data["TFI"] = data["VFI"] + data["SFI"]

# SMI 열을 만들건데 SMA (Area) / (Height / 100) ^ 2
data["SMI"] = data["SMA (Area)"] / (data["Height"] / 100) ** 2
```

```python
from scipy.stats import pearsonr
import pandas as pd
import numpy as np

# 열 선택
columns_of_interest = ["BMI", "VFA (Area)", "SFA (Area)", "VFI", "SFI", "TFI", "SMI"]

# 상관계수와 p-value 저장을 위한 빈 데이터프레임 생성
correlation_results = pd.DataFrame(
    index=columns_of_interest, columns=columns_of_interest
)

# p-value 저장용 데이터프레임
p_values = pd.DataFrame(index=columns_of_interest, columns=columns_of_interest)

# 상관계수와 p-value 계산
for col1 in columns_of_interest:
    for col2 in columns_of_interest:
        if col1 == col2:
            # 자기 자신과의 상관계수는 1, p-value는 0으로 설정
            correlation_results.loc[col1, col2] = 1.0
            p_values.loc[col1, col2] = 0.0
        else:
            r, p = pearsonr(data[col1], data[col2])
            correlation_results.loc[col1, col2] = r
            p_values.loc[col1, col2] = p

# 결과 출력
print("Pearson Correlation Coefficient Matrix:")
print(correlation_results)

print("\nP-value Matrix:")
print(p_values)

# 히트맵 시각화 (상관계수와 p-value)
plt.figure(figsize=(12, 6))

# 상관계수 히트맵
plt.subplot(1, 2, 1)
sns.heatmap(correlation_results.astype(float), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Coefficient Heatmap")

# P-value 히트맵
plt.subplot(1, 2, 2)
sns.heatmap(p_values.astype(float), annot=True, cmap="coolwarm", fmt=".2g", cbar=True)
plt.title("P-value Heatmap")

plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# # 데이터 생성 (실제 데이터로 대체)
# np.random.seed(42)
# data = {
#     "VFI": np.random.uniform(0, 1, 100),
#     "SMI": np.random.uniform(0, 1, 100),
#     "SFI": np.random.uniform(0, 1, 100),
#     "NAFLD Type": np.random.choice(["NAFL", "NASH", "Probable NASH", "Cirrhosis"], 100),
# }
# data_dropped = pd.DataFrame(data)

# Binary label 생성: NASH vs. NOT NASH
# data_dropped.loc[:, "Label"] = (data_dropped["NAFLD Type"] == "NASH").astype(int)

# Feature와 Label 설정
X = data[
    [
        "BMI",
        "VFA (Area)",
        "SFA (Area)",
        "VFI",
        "SFI",
        "TFI",
        "SMI",
    ]
].values
y = data["NASH_label"].values

# 데이터 표준화
scaler = StandardScaler()
X = scaler.fit_transform(X)

# PyTorch 데이터 준비
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1)


# 선형 모델 정의
class LinearModel(nn.Module):
    def __init__(self, input_dim):
        super(LinearModel, self).__init__()
        self.linear = nn.Linear(input_dim, 1, bias=False)  # Weight만 학습

    def forward(self, x):
        return self.linear(x)


# 모델 초기화
model = LinearModel(input_dim=X.shape[1])
# linear model weight 초기화 랜덤으로
model.linear.weight.data = torch.randn_like(model.linear.weight.data)

criterion = nn.BCEWithLogitsLoss()  # Binary Cross-Entropy with Logits
optimizer = optim.Adam(model.parameters(), lr=0.001)
# scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="max", patience=100, verbose=True)

# 모델 학습
num_epochs = 50000
best_auc = 0
best_outputs = None
best_probs = None

for epoch in range(num_epochs):
    # Forward pass
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    # scheduler.step(loss)

    # AUC 계산
    with torch.no_grad():
        probs = torch.sigmoid(outputs).detach().numpy()
        auc = roc_auc_score(y, probs)

    # 최적 AUC와 가중치 저장
    if auc > best_auc:
        best_auc = auc
        best_weights = model.linear.weight.detach().numpy().flatten()
        best_outputs = model(X_tensor)  # 현재 모델의 예측값
        best_probs = torch.sigmoid(best_outputs).detach().numpy()

    # 100회 단위로 로그 출력
    if (epoch + 1) % 100 == 0:
        print(
            f"Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}, AUC: {auc:.4f}"
        )

# 최적의 가중치 출력
optimal_weights = model.linear.weight.detach().numpy().flatten()
print(f"Optimal Weights: {optimal_weights}")
print(f"best Weights: {best_weights}")
print(f"Maximum AUC: {best_auc:.4f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# Device 설정 (GPU 사용 가능하면 CUDA, 아니면 CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Binary label 생성: NASH vs. NOT NASH
# data.loc[:, "Fibrosis Label"] = (data["Fibrosis Stage"] == "High").astype(int)

# Feature와 Label 설정
X = data[
    [
        "BMI",
        "VFA (Area)",
        "SFA (Area)",
        "VFI",
        "SFI",
        "TFI",
        "SMI",
    ]
].values
y = data["Fibrosis Label"].values
# y = data["NASH_label"].values

# 데이터 표준화
scaler = StandardScaler()
X = scaler.fit_transform(X)

# PyTorch 데이터 준비 (device로 이동)
X_tensor = torch.tensor(X, dtype=torch.float32).to(device)
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1).to(device)


# 선형 모델 정의
class LinearModel(nn.Module):
    def __init__(self, input_dim):
        super(LinearModel, self).__init__()
        self.linear = nn.Linear(input_dim, 1, bias=False)  # Weight만 학습

    def forward(self, x):
        return self.linear(x)


learning_rate = 0.0001
# 모델 초기화 (device로 이동)
model = LinearModel(input_dim=X.shape[1]).to(device)
model.linear.weight.data = torch.randn_like(model.linear.weight.data)

criterion = nn.BCEWithLogitsLoss()  # Binary Cross-Entropy with Logits
optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=1e-5)

# Cosine Annealing Scheduler 적용
# scheduler = cosine_annealing_warmup.scheduler.CosineAnnealingWarmupRestarts(
#     optimizer, 100, max_lr=learning_rate, min_lr=0.05, warmup_steps=10, gamma=1.0
# )
scheduler = None

# 모델 학습
num_epochs = 30000
best_auc = 0
best_weights = None
best_probs = None

for epoch in range(num_epochs):
    model.train()

    # Forward pass
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if scheduler is not None:
        scheduler.step()

    # AUC 계산
    model.eval()
    with torch.no_grad():
        probs = torch.sigmoid(outputs).cpu().numpy()  # GPU -> CPU 변환
        auc = roc_auc_score(y, probs)

    # 최적 AUC와 가중치 저장
    if auc > best_auc:
        best_auc = auc
        best_weights = model.linear.weight.detach().cpu().numpy().flatten()
        best_probs = torch.sigmoid(model(X_tensor)).detach().cpu().numpy()

    # 100회 단위로 로그 출력
    if (epoch + 1) % 100 == 0:
        print(
            f"Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}, AUC: {auc:.4f}, LR: {optimizer.param_groups[0]['lr']:.6f}"
        )

# 최적의 가중치 출력
optimal_weights = model.linear.weight.detach().cpu().numpy().flatten()
print(f"Optimal Weights: {optimal_weights}")
print(f"Best Weights: {best_weights}")
print(f"Maximum AUC: {best_auc:.4f}")
```
