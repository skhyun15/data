```python
import pandas as pd

# Load the uploaded Excel file to inspect its structure and contents
file_path = "C:/Users/HOME/Desktop/이혜원 교수님 연구/Deepcatch 연구/Data/Deepcatch 2025-1-14.xlsx"
df = pd.read_excel(file_path, header=2)

# Display sheet names to understand the structure of the Excel file
df.columns
```

```python
# Rename and filter columns
columns_to_extract = {
    "(실명)등록번호": "ID",
    "생년월일": "Birth Date",
    "Fibrosis stage": "Fibrosis Stage",
    "NAFLD Type": "NAFLD Type",
    "사망일자": "Death Date",
    "Last fu_date": "Last F/U",
    "Smoking": "Smoking",
    "LRE": "LRE",
    "LRE_date": "LRE Date",
    "SBP (수축기 혈압) mmHg": "SBP",
    "DBP (이완기 혈압) mmHg": "DBP",
    "AVF (Whole Volume)": "VFA (Volume)",
    "SF (Whole Volume)": "SFA (Volume)",
    "Spleen (Volume)": "Spleen (Volume)",
    "AVF (Single Area)": "VFA (Area)",
    "SF (Single Area)": "SFA (Area)",
    "Muscle (Single Area)": "SMA (Area)",
    "AVF (Single HU)": "VFA (Attenuation)",
    "SF (Single HU)": "SFA (Attenuation)",
    "Muscle (Single HU)": "SMA (Attenuation)",
    "Liver/Spleen HU Ratio": "Liver/Spleen HU",
}

# Select the specified ranges and additional columns
cols_ct_b_to_drinking = df.loc[:, "CT_B date":"BMI"].columns.tolist()
cols_diabetes_to_nephropathy = df.loc[
    :, "당뇨병/당뇨병 전단계 유무":"Nephropathy 유무"
].columns.tolist()

# Combine all column groups
all_columns = (
    list(columns_to_extract.keys())
    + cols_ct_b_to_drinking
    + cols_diabetes_to_nephropathy
)
```

```python
all_columns
```

```python
# Filter the DataFrame to keep only these columns
filtered_df = df[all_columns]

# Rename specified columns
filtered_df = filtered_df.rename(columns=columns_to_extract)

# Create Death and Death Date columns based on the 사망일자 column
filtered_df["Death"] = filtered_df["Death Date"].apply(lambda x: 0 if x == 0 else 1)
filtered_df["Death Date"] = filtered_df["Death Date"].apply(
    lambda x: 0 if x == 0 else x
)
```

```python
# Step 2: Fill missing values
filtered_df["LRE"] = filtered_df["LRE"].fillna(0)
filtered_df["LRE Date"] = filtered_df["LRE Date"].fillna(0)
```

```python
# Step 1: Find unique values in the 'NAFLD Type' column
unique_values = filtered_df["NAFLD Type"].unique()
print("Unique values in 'NAFLD Type':", unique_values)

# Step 2: Define the list of values to filter out
X = [
    "Others (Cancer)",
    "NAFL (+)",
    "Probable NASH (+)",
    "NASH (+)",
    "Missing Biopsy",
]  # Example list of values to remove

# Step 3: Remove rows where 'NAFLD Type' is in the list X
no_cancer_df = filtered_df[~filtered_df["NAFLD Type"].isin(X)]
no_cancer_df = no_cancer_df.reset_index(drop=True)
```

```python
import numpy as np

# Update the 'Fibrosis Stage' column
no_cancer_df["NASH"] = np.where(
    no_cancer_df["NAFLD Type"].isin(["NASH", "Cirrhosis"]), "Yes", "No"
)

# Display the updated DataFrame
print(no_cancer_df["NASH"].value_counts())
```

```python
# Step 1: Find unique values in the 'NAFLD Type' column
unique_values = no_cancer_df["Fibrosis Stage"].unique()
print("Unique values in 'Fibrosis Stage':", unique_values)

import numpy as np

# Update the 'Fibrosis Stage' column
no_cancer_df["Fibrosis Stage"] = np.where(
    no_cancer_df["Fibrosis Stage"].isin([3, 4]), "High", "Low"
)

# Display the updated DataFrame
print(no_cancer_df["Fibrosis Stage"].unique())
```
